## 2. Контрольні питання

- У чому суть задачі знаходження LCS?
Задача полягає в тому, щоб визначити найдовшу підпослідовність символів, яка зустрічається у двох або більше рядках у тому ж порядку, але не обов’язково підряд.
- Які основні методи використовують для знаходження LCS?
Найпоширеніші методи:динамічне програмування,рекурсивний підхід,алгоритм Хіршберга (Hirschberg, не Хаббард),повний перебір.
- Як працює алгоритм динамічного програмування для LCS?
Створюється таблиця, де кожна клітинка відображає довжину спільної підпослідовності між префіксами обох рядків. Значення обчислюються на основі попередніх результатів, залежно від того, чи співпадають символи, чи ні.
- Як працює алгоритм Хіршберга для LCS?
Алгоритм поділяє одну з послідовностей навпіл, а другу — порівнює з кожною частиною, використовуючи динамічне програмування. Одна з половин обробляється у зворотному порядку, щоб знайти точку оптимального поділу, після чого результат об'єднується. Метод зменшує використання пам’яті.

- Які переваги і недоліки мають методи динамічного програмування та Хіршберга?

Динамічне програмування: просте у реалізації, але споживає багато пам’яті при великих рядках.

Хіршберг: ефективніший по пам’яті (O(n) замість O(n×m)), але складніший у реалізації і часто повільніший через рекурсивну природу.
- Де застосовується задача LCS на практиці?
Пошук плагіату,Автоматична перевірка орфографії,Аналіз версій текстів (diff),Визначення схожості між ДНК або текстами,Порівняння файлів у системах контролю версій.
